```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = sns.load_dataset('mpg')
df
```

```python
sns.relplot(x='acceleration', y='mpg', col='origin', data=df)
print('''Carros produzidos nos EUA e no Japão possuem uma correlação positiva, onde o consumo de combustível tende a aumentar com a aceleração do carro.
Já na Europa não existe tal correlação.''')
```

```python
sns.scatterplot(x='acceleration', y='mpg', hue='model_year', data=df)
print('''Carros mais antigos tendem a ter um consumo de combustível menor, assim como a aceleração. 
Carros mais novos tendem a ter um consumo de combustível maior.''')
```
