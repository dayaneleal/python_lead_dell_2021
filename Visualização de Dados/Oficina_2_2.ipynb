```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
videogames = pd.read_csv('https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro04/oficina/arquivos/dadosoficina04.txt')
videogames
```

Proporção de vendas global entre gêneros de jogos da empresa Nintendo

```python
nintendo = videogames[videogames['Publisher'] == 'Nintendo']
generos = nintendo.groupby(by='Genre').sum()
generos
```

```python
generos = generos.sort_values(by='Global_Sales', ascending = False)
porcentagens =  generos['Global_Sales']*100.0/generos['Global_Sales'].sum()
plt.figure(figsize=(15,5))
plt.bar(generos.index,porcentagens)
plt.title('Proporção de vendas de jogos por gênero da Nintendo')
plt.style.use('seaborn-darkgrid')
```

Proporção de quantidade de jogos por plataforma a partir do ano de 2015.

```python
ano = videogames[videogames['Year'] > 2015]
quantidade_por_plataforma = ano.groupby(by = 'Platform').count()
porcentagens =  quantidade_por_plataforma['Rank']*100.0/quantidade_por_plataforma['Rank'].sum()
plt.title('Quantidade de jogos por plataforma a partir do ano de 2015')
plt.bar(quantidade_por_plataforma.index,porcentagens)
```
