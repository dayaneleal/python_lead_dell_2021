```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
plt.figure(figsize=(10,8))
sns.boxenplot(x="embark_town", y="fare",  data=titanic)
```

```python
sns.catplot(x="embark_town", y="fare", hue="sex", col="survived", kind="violin", split = True, data=titanic)
```

No houve grande diferena no preo pago na passagem dos passageiros sobreviventes e no sobreviventes que embarcaram em Southampton e Queenstown. Entretanto, os passageiros que pagaram mais de 300 e embarcaram em Cherbourg sobreviveram ao naufgio.
