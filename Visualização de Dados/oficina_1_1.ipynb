```python
import pandas as pd
```

```python
df = pd.read_csv('drinks.txt')
```

```python
df
```

# Qual a média e desvio padrão do consumo de cerveja, licor, vinho e total de álcool?

```python
print(f'''- Média do consumo mundial -\n 
Cerveja: {df['beer_servings'].mean()} 
Vinho: {df['wine_servings'].mean()} 
Licor: {df['spirit_servings'].mean()}
Média do total de álcool: {df['total_litres_of_pure_alcohol'].mean()}
''')
```

```python
print(f'''- Desvio padrão do consumo mundial -\n 
Cerveja: {df['beer_servings'].std()} 
Vinho: {df['wine_servings'].std()} 
Licor: {df['spirit_servings'].std()}
Média do total de álcool: {df['total_litres_of_pure_alcohol'].std()}
''')
```

# Qual o consumo de licor na França?

```python
france = df[df['country'] == 'France'] 
print(f"O consumo é de Licor na França é de {france['spirit_servings'].to_string(index=False)}.")
```

# Quais os 5 países nos quais se consome mais vinho?

```python
df[['country','wine_servings']].sort_values(by='wine_servings',ascending = False).head()
```

# Quais os países nos quais o consumo de cerveja está acima da média?

```python
paises = df[df['beer_servings'] > df['beer_servings'].mean()]
paises[['country','beer_servings']].sort_values(by = 'beer_servings', ascending = False)
```

# Quais países nos quais o consumo total de álcool está acima do Brasil, em ordem do menor para o maior?

```python
brasil = df[df['country'] == 'Brazil']
total_consumo_brasil = brasil['total_litres_of_pure_alcohol'].values[0]

distribuicao_paises = df[df['total_litres_of_pure_alcohol'] > total_consumo_brasil]

distribuicao_paises[['country','total_litres_of_pure_alcohol']].sort_values(by = 'total_litres_of_pure_alcohol')
```
