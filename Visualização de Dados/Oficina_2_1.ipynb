```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
filmes_uri = 'https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro03/oficina/arquivos/dadosoficina03.txt'
filmes = pd.read_json(filmes_uri)
filmes
```

# Orçamento de todos os filmes do conjunto de dados

```python

plt.figure(figsize=(5,15))
sns.set(style="darkgrid")
sns.barplot(x='Budget_M', y='Film', data=filmes, ci=None)
plt.ylabel('Filme')
plt.xlabel('Orçamento')
```

```python
print("O filme com o maior orçamento foi Pirates of the Caribbean: At World's End.")
```

```python
print("O filme com o menor orçamento foi The Twilight Saga: New Moon.")
```

# Bilheteria de todos os filmes do conjunto de dados

```python
plt.figure(figsize=(8,15))
sns.set(style="darkgrid")
sns.barplot(x='Worldwide_Gross_M', y='Film', data=filmes, ci=None)
plt.xlabel("Bilheteria Mundial")
```

```python
print("O filme com a maior bilheteria foi Avatar.")
```

```python
filme_min = filmes['Film'][filmes['Worldwide_Gross_M'] == filmes['Worldwide_Gross_M'].min()].to_string(index = False)
print(f"O filme com a menor bilheteria foi {filme_min}.")
```

#  Quantidade de filmes por gênero em cada ano

```python
plt.figure(figsize=(15,8))
sns.countplot(x='Year', hue='Genre', data=filmes)
plt.ylabel("Contagem")
plt.xlabel("Ano")
plt.legend(loc='upper right', title='Gênero')
```

```python
print('Os anos com menos filmes de ação foram 2008 e 2010');
```

```python
print('Os anos com menos filmes de animação foram 2008 e 2011');
```

# Média de orçamento de filmes por gênero em cada ano

```python
plt.figure(figsize=(15,7))
sns.barplot(x='Year', y='Budget_M', hue='Genre',ci=None ,data=filmes)
plt.ylabel('Orçamento')
plt.xlabel('Ano')
plt.legend(loc='upper right', title="Gênero")
```

```python
print('O ano em que os filmes de ação tiveram maior média de orçamento foi 2007.' )
```

```python
print('O ano em que os filmes de aimação tiveram maior média de orçamento foi 2008.' )
```
