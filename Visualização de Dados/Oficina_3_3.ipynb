```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
uri = 'https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro07/oficina/arquivos/dadosoficina07.txt'
df = pd.read_csv(uri)
df    
```

```python

plt.figure(figsize=(20,5))
plt.title('Variação das vendas mensais de shampoo durante um período de 3 anos')
sns.lineplot(x="Month", y="Sales", data=df)
```

```python
print('''Nos últimos 3 anos houve uma tendência de aumento na venda do produto. 
Pela observação do gráfico não se verifica nenhuma periodicidade.''')
```
