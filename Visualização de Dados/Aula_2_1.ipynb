```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
categorias = ['A','B','C','D']
valores = [10,20,15,5]
plt.bar(categorias, valores)
plt.xlabel('Grupos')
plt.ylabel('Tamanho')
plt.title('Meu primeiro gráfico de barras')
plt.show()
```

```python
filmes_uri = 'https://raw.githubusercontent.com/emanueles/datavis-course/master/assets/files/observable/movies.json'
filmes = pd.read_json(filmes_uri)

plt.figure(figsize=(15,8))
plt.title('Filmes por gênero em cada ano')
##Genero será mapeado como cor para o gráfico de barras agrupado
sns.barplot(x="Year", y="Worldwide_Gross_M", palette="rocket", edgecolor="1", ci=None, hue='Genre',data=filmes)
plt.show()
```
